# Simulation Results

This Chapter is intended to analyse and compare the newly derived controller to the pseudoinversed Jacobian approach, which is common in literature. This test is in combination with the dual quaternion interpolation scheme. These tests are based on the robot welding example and interactive showcases of the predictive task directional manipulablity optimization of the robot configuration.

The depicted motion runs through the wrist and shoulder singularities as these are common during the welding process. The motion commands are simplified to easily show the effect of the predictive controller vs the standard approach.

```python
robot_type = "weld"

fk = ForwardKinematics(robot_type)

q1 = np.array([deg2rad(0), deg2rad(40), 0, deg2rad(70), 0, deg2rad(40), 0])

# orientations for first trajectory
quat0 = Quaternion.fromAxisAngle(np.pi*1.0, np.array([0,1,0]))

# first line segments
x1 = DualQuaternion.fromQuatPos(quat0, np.array([0.7, 0, 0.3]))
x2 = DualQuaternion.fromQuatPos(quat0, np.array([0.9, 0.0, 0.2]))
x3 = DualQuaternion.fromQuatPos(quat0, np.array([0.9, 0.3, 0.2]))
x4 = DualQuaternion.fromQuatPos(quat0, np.array([0.9, -0.3, 0.2]))

x5 = DualQuaternion.fromQuatPos(quat0, np.array([0.95, -0.3, 0.3]))
x6= DualQuaternion.fromQuatPos(quat0, np.array([0.95, 0.3, 0.3]))
x7 = DualQuaternion.fromQuatPos(quat0, np.array([0.9, -0.3, 0.2]))

task_list = np.array([MoveJoint(q1, 4), MoveLinear(fk.getFK(q1), x1, 4), MoveLinear(x1, x2, 7), MoveLinear(x2, x3, 7), 
                      MoveLinear(x3, x4, 7), MoveLinear(x4, x5, 7), MoveLinear(x5, x6, 7), MoveLinear(x6, x7, 7), WaitTime(2)])

sim = Simulation(task_list, robot_type, method = "qp")

sim.start()
```

```python
robot_type = "weld"

fk = ForwardKinematics(robot_type)

q1 = np.array([deg2rad(0), deg2rad(40), 0, deg2rad(70), 0, deg2rad(40), 0])

# orientations for first trajectory
quat0 = Quaternion.fromAxisAngle(np.pi*1.0, np.array([0,1,0]))
quat1 = Quaternion.fromAxisAngle(np.pi*0.5, np.array([0,0,1]))

# first line segments
x1 = DualQuaternion.fromQuatPos(quat0, np.array([0.7, 0, 0.3]))
x2 = DualQuaternion.fromQuatPos(quat1*quat0, np.array([0.9, 0.0, 0.2]))
x3 = DualQuaternion.fromQuatPos(quat0, np.array([0.9, 0.3, 0.2]))
x4 = DualQuaternion.fromQuatPos(quat0, np.array([0.9, -0.3, 0.2]))

x5 = DualQuaternion.fromQuatPos(quat0, np.array([0.95, -0.3, 0.3]))
x6= DualQuaternion.fromQuatPos(quat0, np.array([0.95, 0.3, 0.3]))
x7 = DualQuaternion.fromQuatPos(quat0, np.array([0.9, -0.3, 0.2]))

task_list = np.array([MoveJoint(q1, 4), MoveLinear(fk.getFK(q1), x1, 4), MoveLinear(x1, x2, 7), MoveLinear(x2, x3, 7), 
                      MoveLinear(x3, x4, 7), MoveLinear(x4, x5, 7), MoveLinear(x5, x6, 7), MoveLinear(x6, x7, 7), WaitTime(2)])

sim = Simulation(task_list, robot_type, method = "qp")

sim.start()
```

# Double Cover of $\mathcal{SE}(3)$ 

The double cover of $\mathcal{SE}(3)$ form the dual quaternion set $\mathbb{H}$ is an important advantage over $\mathcal{SE}(3)$. As cartesian cobot tasks are often set up by the hand guiding mode, the user intuitively thinks that the cobot moves as taught. As shown in Chapter 3.0, this property can be leveraged to define directionality in the orientation interpolation. The following code blocks will show this feature in the simulation environment. First, two joint configuration are defined, as would be the case for hand guiding the robot into the desired cartesian position from its home position. From these two joint configuration $q1$ and $q2$, the forward kinematics are computed to get the dual quaternionic pose of the end-effector. 

```python
robot_type = "normal"
fk = ForwardKinematics(robot_type)

q1 = np.array([deg2rad(-60), deg2rad(20), 0, deg2rad(70), 0, deg2rad(90), deg2rad(160)])
q2 = np.array([deg2rad(0), deg2rad(20), 0, deg2rad(70), 0, deg2rad(90), deg2rad(-120)])

x1 = fk.getFK(q1)
x2 = fk.getFK(q2)

T2 = x2.asTransformation()

x2_1  = DualQuaternion.fromTransformation(T2)
print(x2_1)
print(x2)

angle1 = (x1.real*x2.real.inverse()).getAngle()
angle2 = (x1.real*x2_1.real.inverse()).getAngle()
print("angle between x1 and x2: ", angle1)
print("angle between x1 and x2_1: ", angle2)
```

To simulate what a homogeneous matrix based forward kinematics representation would yield, the second dual quaternion $x2$ was converted to a homoegeneous transformation matrix. This matrix represents the same pose as the original matrix but lacks the double cover property. This can be seen by the mapping: $\mathcal{SE}(3) \to \mathbb{H}$. Here, in this example the result will be the antipodal $-x2$ as can be seen by the print. The resulting interpolation will then run into joint limitations as the MoveLinear interpolation from joint configuration $q1$ to $x2\_1$ will take the shortest path, instead of the long path which is natural and intuitive to the user. This is also confirmed by the computed angle between $x1$, which is the respective pose to joint configuration $q1$ and the dual quaternionic representations of the second pose, here the angle between $x1$ and $x2$ is greater than $\pi$ and clearly represents the long path interpolation, where as the angle between $x1$ and $x2\_1$ is small.

Example 1: Interpolation with the original dual quaternion

```python
task_list = np.array([MoveJoint(q1, 4), MoveLinear(fk.getFK(q1), x2, 4), WaitTime(1)])

sim = Simulation(task_list, robot_type, method = "classic")

sim.start()

plotJointAngles(sim.time_list, sim.q_list)
```

Example 2: Interpolation with the antipodal dual quaternion $x2\_1$

```python
task_list = np.array([MoveJoint(q1, 4), MoveLinear(fk.getFK(q1), x2_1, 4), WaitTime(1)])

sim = Simulation(task_list, robot_type, method = "classic")

sim.start()

plotJointAngles(sim.time_list, sim.q_list)
```

From the plots it is quite obvious that the incorrect direction of the orientation interpolation will lead to an joint limit violation in joint 7. This is often the case of orientation angles exceed 180 degrees cumulatively. For orientations below 180 degrees, the direction of the orientation unit quaternion retrieved from the homogenoues transformation matrices match the ones from the unit dual quaternions. Yet, this is a clear advantage of the dual quaternionic kinematics representation in motion-planning. Different comparisons can be compared by the joint angles in $q2$ from the joint configuration definition above.
